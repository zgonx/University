```python
import numpy as np
import matplotlib.pyplot as plt
def Newton( n, k ):
 
    val = 1
    for i in range( 1, k+1 ):
        val = val * ( n - i + 1 ) / i
    return val

def Bernstein(n, i, t):
    B=Newton(n, i)*(t**i)*((1.0-t)**(n-i))
    return B

def pBezier(points):
    data=[]
    data.append([])
    data.append([])
    n = len(points)-1
    ll = 100
    for i in range( ll+1):
        t=i/ll
        
        px=0
        py=0
        for j in range(n+1):
            bb=Bernstein(n, j,t)
            #print(bb)
            px+=points[j][0]*bb
            py+=points[j][1]*bb
        print(px, py)
        data[0].append(px)
        data[1].append(py)        
    return data
```

```python
data=pBezier([[0,40],[0, 0], [200,50], [-150,100], [50, 200],[45,150]])
# data2=pBezier([[0,40],[0,0],[100,50],[-70,100],[10,100],[60,100],[0,150],[0,125]])
# data2=pBezier([[40,40],[0,0],[0,40],[]])
#data=np.ones((500,500,3))
a=plt.plot(data[0], data[1])
# b=plt.plot(data2[0], data2[1])

plt.show()
```
