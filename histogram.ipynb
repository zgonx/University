```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
im = Image.open("Koloseum.png")
data = np.array(im)
# print(data)
black = 0
red = 0
a = np.arange(256)

def gamma(data, gamma):
    for i in range(len(data)):
        for j in range(len(data[0])):
            for k in range(3):
                data[i][j][k] = 255 * (data[i][j][k] / 255) ** gamma

ran = [el for el in range(0,256)]

for pixel in im.getdata():
    if pixel == (0,0,1):
        black += 1
    else:
        red += 1
print(black,red)
histR = np.zeros(256)
histG = np.zeros(256)
histB = np.zeros(256)
for i in data:
    for j in i:
        histR[j[0]] +=1
        histG[j[1]] +=1
        histB[j[2]] +=1

plt.bar(range(256), histR, color='red', width = 1)
plt.bar(range(256), histG, color='green', width = 1)
plt.bar(range(256), histB, color='blue', width = 1)

```

```python
img = Image.open("Koloseum.png")


RGB = np.array(img)
gamma(RGB, 0.5)

newImg = Image.fromarray(RGB, 'RGB')

display(newImg)
```

```python
plt.clf()
data = np.array(newImg)

histR = np.zeros(256)
histG = np.zeros(256)
histB = np.zeros(256)
for i in data:
    for j in i:
        histR[j[0]] +=1
        histG[j[1]] +=1
        histB[j[2]] +=1

plt.bar(range(256), histR, color='red', width = 1)
plt.bar(range(256), histG, color='green', width = 1)
plt.bar(range(256), histB, color='blue', width = 1)
```
